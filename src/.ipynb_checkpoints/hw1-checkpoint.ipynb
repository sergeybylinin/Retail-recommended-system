```python
class WorkFllow:
    
    def __init__(self, recommended_list, bought_list, 
                 prices_recommended = None, 
                 prices_bought = None,
                 k = None):
        self.recommended_list = recommended_list
        self.bought_list = bought_list
        self.prices_recommended = prices_recommended
        self.prices_bought = prices_bought
        self.k = k
        
    @staticmethod
    def _hit_rate(recommended_list, bought_list):
        recommended_list = np.array(recommended_list)
        bought_list = np.array(bought_list)        
        flags = np.isin(bought_list, recommended_list)
        return (flags.sum() > 0) * 1
    
    @staticmethod
    def _hit_rata_at_k(recommended_list, bought_list, k):
        return WorkFllow._hit_rate(recommended_list[:k], bought_list)
    
    @staticmethod
    def _precision(recommended_list, bought_list):
        recommended_list = np.array(recommended_list)
        bought_list = np.array(bought_list)        
        flags = np.isin(bought_list, recommended_list)
        return flags.sum() * 1
    
    @staticmethod
    def _precision_at_k(recommended_list, bought_list, k):
        return WorkFllow._precision(recommended_list[:k], bought_list)
    
    @staticmethod
    def _money_precision_at_k(recommended_list, bought_list, 
                              prices_recommended, k):
        recommended_list = np.array(recommended_list)[:k]
        bought_list = np.array(bought_list)        
        prices_recommended = np.array(prices_recommended)[:k]
        flags = np.isin(recommended_list, bought_list)
        return np.dot(flags, prices_recommended).sum() / prices_recommended.sum()
    
    @staticmethod
    def _recall(recommended_list, bought_list):
        recommended_list = np.array(recommended_list)
        bought_list = np.array(bought_list)        
        flags = np.isin(bought_list, recommended_list)
        return flags.sum / len(bought_list)
    
    @staticmethod
    def _recall_at_k(recommended_list, bought_list, k):
        return WorkFllow._recall(recommended_list[:k], bought_list)
    
    @staticmethod
    def _money_recall_at_k(recommended_list, bought_list, 
                           prices_recommended, prices_bought, k):
        recommended_list = np.array(recommended_list)[:k]
        bought_list = np.array(bought_list)
        prices_recommended = np.array(prices_recommended)[:k]
        prices_bought = np.array(prices_bought)
        flags = np.isin(recommended_list, bought_list)
        return np.dot(flags, prices_recommended).sum() / prices_bought.sum()
    
    @staticmethod
    def _ap_k(recommended_list, bought_list, k):
        recommended_list = np.array(recommended_list)
        bought_list = np.array(bought_list)
        flags = np.isin(recommended_list, bought_list)
        
        if sum(falags) == 0: 
            return 0
        
        sum_ = 0
        for i in range(k-1):
            if flags[i] == True:
                p_k = WorkFllow._precision_at_k(
                    recommended_list, bought_list, k=i+1)
                sum_ += p_k

        return sum_ / sum(flags)
    
    @staticmethod
    def _reciprocal_rank(recommended_list, bought_list):
        recommended_list = np.array(recommended_list)
        bought_list = np.array(bought_list)
        flags = np.isin(bought_list, recommended_list)
        return np.mean(1 / flags.nonzero()[0][0] + 1) if flags.size else 0
    
    
    def hit_rate_at_k(self):
        assert self.k is not None, 'No k entered!'
        return WorkFllow._hit_rata_at_k(self.recommended_list,
                                        self.bought_list, 
                                        self.k)
    
    def money_precision_at_k(self):
        assert self.k is not None, 'No k entered!'
        assert self.prices_recommended is not None, 'No prices_recommendet entered!'
        return WorkFllow._money_precision_at_k(self.recommended_list,
                                               self.bought_list,
                                               self.prices_recommended,
                                               self.k)
    
    def recall_at_k(self):
        assert self.k is not None, 'No k entered!'
        return WorkFllow._recall_at_k(self.recommended_list,
                                      self.bought_list,
                                      self.k)
        
    def money_recall_at_k(serl):
        assert self.k is not None, 'No k entered!'
        assert self.prices_recommended is not None, 'No prices_recommendet entered!'
        assert self.prices_bought is not None, 'No prices_bought entered!'
        return WorkFllow._money_recall_at_k(self.recommended_list,
                                            self.bought_list,
                                            self.prices_recommended,
                                            self.prices_bought,
                                            self.k)
    
    def ap_k(self):
        assert self.k is not None, 'No k entered!'
        return WorkFllow._ap_k(self.recommended_list,
                               self.bought_list,
                               self.k)
    
    def reciprocal_rank(self):
        return WorkFllow._reciprocal_rank(self.recommended_list,
                                          self.bought_list)
```
